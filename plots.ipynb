```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[4,5,5]
plt.plot(x,y)
plt.title("info")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
import numpy as np
y=np.array([35,25,25,15])
mylabels=["apples","bananas","cherry","watermelon"]
mycolors=["r","hotpink","b","#4CAF50"]
n=([0.34,0.5,0.1,0.3])
plt.pie(y,labels=mylabels,colors=mycolors,explode=n,shadow=True)
plt.show()

```

```python
x=[1,2,3,4]
y=[4,5,6,8]
plt.xlabel("x axis",fontsize=20)
plt.ylabel("y axis",fontsize=20)
plt.title("line graph",fontsize=10)
mycolors=["r","hotpink","b","#4CAF50"]
plt.bar(x,y,color=mycolors)
```

```python
x=[1,2,3,4]
y=[4,5,6,8]
plt.xlabel("x axis",fontsize=20)
plt.ylabel("y axis",fontsize=20)
plt.title("line graph",fontsize=10)
mycolors=["r","hotpink","b","#4CAF50"]
plt.barh(x,y,color=mycolors)
```

```python
x=[1,2,3,4]
y=[4,5,6,8]
plt.xlabel("x axis",fontsize=20)
plt.ylabel("y axis",fontsize=20)
sizes=[500,200,50,40]
plt.title("line graph",fontsize=10)
mycolors=["r","hotpink","b","#4CAF50"]
plt.scatter(x,y,color=mycolors,marker="*",edgecolor="r",s=sizes,alpha=0.8)
plt.show()
```
